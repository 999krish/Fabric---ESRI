### **Key Indices Used**
‚úÖ **Normalized Burn Ratio (NBR)** ‚Äì Highlights burned areas using near-infrared (NIR) and shortwave infrared (SWIR) reflectance differences.  
‚úÖ **Normalized Difference Vegetation Index (NDVI)** ‚Äì Measures vegetation health, useful for pre- and post-fire comparisons.  
‚úÖ **QA Pixel Data** ‚Äì Identifies cloud cover, shadows, and sensor anomalies in Landsat images to improve data quality.

## **2Ô∏è‚É£ Spectral Band Analysis for Fire Detection**
To further enhance our fire analysis, we examined **Landsat Bands 5, 6, and 7**, which are particularly effective for detecting burned areas and vegetation stress.

| **Band**  | **Name**                        | **Wavelength (¬µm)**  | **Use in Fire Analysis** |
|-----------|--------------------------------|----------------------|--------------------------|
| **Band 5** | Near-Infrared (NIR)            | 0.85 ‚Äì 0.88 ¬µm       | Differentiates healthy vegetation from burned areas. |
| **Band 6** | Shortwave Infrared (SWIR 1)    | 1.57 ‚Äì 1.65 ¬µm       | Detects vegetation moisture loss and highlights burned areas. |
| **Band 7** | Shortwave Infrared (SWIR 2)    | 2.11 ‚Äì 2.29 ¬µm       | Identifies fire scars, heat signatures, and post-fire soil reflectance. |

```python
def load_tif(file_path):
    """Load a raster TIFF file from Microsoft Fabric OneLake"""
    with fsspec.open(file_path, mode="rb", filesystem="abfs") as f:
        with rasterio.open(f) as src:
            data = src.read(1)  # Read first band
            meta = src.meta  # Get metadata
            print(f"Loaded: {file_path.split('/')[-1]}")
            print("Metadata:", meta)
            return data, meta

# Load all NBR and QA_PIXEL files
nbr_data = {key: load_tif(nbr_paths[key]) for key in nbr_paths}
qa_data = {key: load_tif(qa_paths[key]) for key in qa_paths}
```

```python
def apply_cloud_mask(nbr, qa):
    """Mask out clouds using the QA_PIXEL band"""
    # Extract the first element if it's a tuple
    if isinstance(nbr, tuple):
        nbr = nbr[0]
    if isinstance(qa, tuple):
        qa = qa[0]

    # Apply cloud mask using bitwise operation
    cloud_mask = (qa & 0b110000000000) == 0  # Mask clouds (bits 12 and 13)
    return np.where(cloud_mask, nbr, np.nan)  # Replace clouds with NaN
```

```python
def apply_cloud_mask(nbr, qa):
    """Mask out only high-confidence clouds (bits 12-13 == 2 or 3)"""
    cloud_mask = ((qa >> 12) & 3) >= 2  # Only mask 2 (medium confidence) and 3 (high confidence)
    return np.where(cloud_mask, np.nan, nbr)  # Replace cloud pixels with NaN

# Apply the updated masking function
nbr_cleaned["NBR_1_Clean"] = apply_cloud_mask(nbr_data["NBR_1"], qa_data["QA_1"])
nbr_cleaned["NBR_2_Clean"] = apply_cloud_mask(nbr_data["NBR_2"], qa_data["QA_2"])

# Visualize again
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))

# Plot cloud-masked Dataset 1
plt.subplot(2, 1, 1)
plt.imshow(nbr_cleaned["NBR_1_Clean"], cmap="RdYlGn", vmin=-1, vmax=1)
plt.colorbar(label="NBR Value")
plt.title("Cloud-Free NBR Dataset 1 (Fixed Masking)")

# Plot cloud-masked Dataset 2
plt.subplot(2, 1, 2)
plt.imshow(nbr_cleaned["NBR_2_Clean"], cmap="RdYlGn", vmin=-1, vmax=1)
plt.colorbar(label="NBR Value")
plt.title("Cloud-Free NBR Dataset 2 (Fixed Masking)")

plt.show()


```
